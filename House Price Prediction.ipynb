```python
from sklearn.linear_model import LinearRegression
x_train,y_train=train_data.drop(['Price'],axis=1),train_data['Price']
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
test_data=x_test.join(y_test)
test_data['Total Rooms']=test_data['number of bedrooms']+test_data['number of bathrooms']
test_data=test_data.drop('Date',axis=1)
test_data=test_data.drop('id',axis=1)
test_data=test_data.drop('Lattitude',axis=1)
test_data=test_data.drop('Longitude',axis=1)
test_data=test_data.drop('Number of schools nearby',axis=1)
test_data=test_data.drop('Distance from the airport',axis=1)
```

```python
x_test,y_test=test_data.drop(['Price'],axis=1),test_data['Price']
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor()
param_grid={
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth":[None,4,8]}
grid_search=GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(x_test,y_test)
```
